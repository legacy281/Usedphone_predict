```python
def histo(str):
    plt.subplot(1, 2, 1)
    # Vẽ histogram
    plt.hist(dfsmall_clean[str], bins=20)

    # Cấu hình định dạng và nhãn
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(str)

    plt.subplot(1, 2, 2)
    # Vẽ histogram
    plt.hist(dfbig_clean[str], bins=20)

    # Cấu hình định dạng và nhãn
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('big '+str)
    # Hiển thị biểu đồ
    plt.show()
```

```python
histo('price')
```

```python
histo("pin")
```

```python
def bar(str):
    plt.figure(figsize = (15, 5))
    plt.rcParams["figure.autolayout"] = True

    counts = dfsmall_clean[str].value_counts()

    plt.subplot(1, 2, 1)
    sns.barplot(x = counts.index, y = counts.values, alpha = 0.8)
    plt.title('Distribution of '+str,fontsize=15 ,fontweight ='bold')
    plt.ylabel('Counts', fontsize = 10)
    plt.xlabel('Value', fontsize = 10)
    counts = dfbig_clean[str].value_counts()
    plt.subplot(1, 2, 2)
    sns.barplot(x = counts.index, y = counts.values, alpha = 0.8)
    plt.title('Distribution of big '+str,fontsize=15 ,fontweight ='bold')
    plt.ylabel('Counts', fontsize = 10)
    plt.xlabel('Value', fontsize = 10)
    plt.tight_layout()
    plt.show()
```

```python
def pie(str,lb0,lb1):
    labels_small = ['1: '+lb1, '0: '+lb0]
    counts = dfsmall_clean[str].value_counts()
    plt.subplot(1, 2, 1)
    plt.pie(counts.values, labels=labels_small, autopct='%1.1f%%')
    plt.title('Distribution of '+str, fontsize=15, fontweight='bold')
    
    counts = dfbig_clean[str].value_counts()
    plt.subplot(1, 2, 2)
    plt.pie(counts.values, labels=labels_small, autopct='%1.1f%%')
    plt.title('Distribution of '+str, fontsize=15, fontweight='bold')
    plt.show()
```

```python
pie('result','ca nhan','chuyen nghiep')
```

```python
pie('condition','qua sua chua','chua sua chua')
```

```python
pie('preserve','het bao hanh','con bao hanh')
```

```python
pie('brand','samsung','iphone')
```

```python
bar('year')
bar('ram')
bar('storage')
bar('display')
bar('camera')
```
